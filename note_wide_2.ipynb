```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_excel("./combined_wide_format.xlsx")
# bunu yeni oluşturduk 
pd.set_option('display.max_columns', 350)
pd.set_option("display.max_rows",400)
```

```python
# beklediğimden daha fazla çıktı daha detaylı inceleyelim bir sorun var gibi 
df_columns = df.columns
columns_name  = [cols.split(".")[0] for cols in df_columns ]
df.columns = columns_name
df.columns
```

```python
#şimdi tekrar bakalım tekrar eden sütun var mı ? 
duplicated_columns = df.columns.duplicated()
duplicated_columns_name  = df.columns[duplicated_columns]
print(duplicated_columns_name)
```

```python
new_df = df[duplicated_columns_name]
new_df.head(10)
```
