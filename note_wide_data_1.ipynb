# Boş olan sütunlar temizlendi 
Sonuç olarak boş sütun kalmamış  fakat çoğunluğu boş olan satırlar ve kopyalanmış satırlar var. 
## ToDO

*bir yüzde belirle bu yüzdede boş olan satırları bul 

```python
new_columns = new_wide_df.columns
new_columns = [cols.split(".")[0] for cols in new_columns]

#yeni sütun adlarını df atama 
# Yeni sütun adlarını DataFrame'e ata
new_wide_df.columns = new_columns

# Sonuç kontrolü
print("Güncellenmiş Sütun Adları:")
display(new_wide_df.head(5))
```

```python
# şimdi satır adlarındaki düzenlemeleri yaptık aynı isimde satır adları var mı bunu inceleyelim 
duplicated_columns = new_wide_df.columns.duplicated()
duplicate_names = new_wide_df.columns[duplicated_columns]
# Tekrar eden sütun isimlerini yazdır
print("Tekrar Eden Sütun İsimleri:")
print(duplicate_names)
#Tekrar eden sütun sayısı yazdırılır 
print("Tekrar eden sütun sayısı : ")
len(duplicate_names)
```

## Tekrar eden sütünlar aynı mı 
Tekrar eden sütun isimlerini bulduk şimdi bu sütünların içeriklerinin aynı olup olmadığını kıyaslamamız gerekiyor
Bunu yapmak için yöntemler araştırmak lazım 

```python
#Sadece tekrar eden sütunların olduğu bir df oluşturalım 
df_deneme = new_wide_df[duplicate_names]
df_deneme.head(10)
    
```
