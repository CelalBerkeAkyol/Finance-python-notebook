# Karşılaştırma 
Elimde farklı sektörlerden farklı şirketler var bunları karşılaştırarak çakışmayı sağlayan şeyin ne olduğunu anlamaya çalışalım 


```python
for comp in company_list:
    # Şirketlerin sütun isimlerini alalım ve temizleyelim
    comp_columns = comp.columns
    columns_name = [cols.strip().lower().split(".")[0] for cols in comp_columns]
    comp.columns = columns_name

    # Tekrarlanan sütunları bulalım
    duplicated_columns = comp.columns.duplicated()
    duplicated_columns_name = comp.columns[duplicated_columns]

    # Tekrarlanan sütun isimlerini yazdıralım
    print("\n\n\nTekrarlanan sütun isimleri: \n\n\n", duplicated_columns_name)
```

Orjinal kaynağı import edelim ve inceleyelim 


```python
OG_AEFES = pd.read_excel("./uzun/AEFES.xlsx")
```

```python
OG_AEFES.head(10)
```

```python
Kalemler = OG_AEFES["Kalem"].to_list()
new_kalemler = [kalem.strip().lower() for kalem in Kalemler ]
OG_AEFES["Kalem"] = new_kalemler
# Kontrol için sütunu yazdır
print(OG_AEFES["Kalem"])
```

```python
#çoğaltılmış var mı ona bakalım 

duplicates_kalem = {kalem for kalem in new_kalemler if new_kalemler.count(kalem) > 1}
print("Tekrarlanan elemanlar:", duplicates_kalem)
```

```python
#Tekrar eden verileri bir görüntüleyelim bakalım. Bunun için tekrar eden satırları bulmamız laızm  

for kalem in duplicates_kalem : 
    filtered_rows = OG_AEFES[OG_AEFES["Kalem"] == kalem]
    # Sonuçları görüntüle
    display(filtered_rows)





```
